```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

X,y = make_blobs(n_samples=150,
                 n_features=2,
                 centers=3,
                 cluster_std=0.5,
                 shuffle=True,
                 random_state=0)
```

```python
def Kmeans(k, x, max_iter=100):
    # 设置初始中心点
    center_point0 = np.array([x[i] for i in range(0, k)])
    # 初始化记录距离的矩阵 k*x
    dist = np.zeros((k, x.shape[0]))
    
    for a in range(0, max_iter):
        # 计算每点到各中心点的距离
        for i in range(0, k):
            dist[i] = np.array([j for j in np.sum((x-center_point0[i])**2,axis=1)])
        # 找到每列最小值的索引
        mindist = np.argmin(dist,0)
        # 更新中心点
        center_point1 = np.array([np.mean(x[mindist==i], 0) for i in range(0, k)])
        # 退出条件
        if np.all(center_point0==center_point1):
            break
        # 继续迭代
        center_point0=center_point1
    
    return center_point1
    
center_p = Kmeans(3, X)    
    
plt.scatter(X[:, 0],X[:, 1])
plt.scatter(center_p[:, 0], center_p[:, 1], c='red')
```
