```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
X,y = make_blobs(n_samples=150,
                 n_features=2,
                 centers=3,
                 cluster_std=0.5,
                 shuffle=True,
                 random_state=0)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
def ou_distance(x,y):
    return np.sqrt(sum(np.square(x - y)))
```

```python
def random_num(dataset,k):
    n = np.shape(dataset)[1]
    cent_point = np.mat(np.zeros((k,n)))
    for j in range(n):
        minJ = min(dataset[:,j])
        rangeJ = float(max(dataset[:,j]) - minJ)
        cent_point[:,j] = minJ + rangeJ * np.random.rand(k,1)
    return cent_point
```
