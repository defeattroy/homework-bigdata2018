```python
from sklearn.datasets import make_blobs
import math
import random
import numpy as np
import matplotlib.pyplot as plt #导入所需要的包

X,y = make_blobs(n_samples=150,
                 n_features=2,
                 centers=3,
                 cluster_std=0.5,
                 shuffle=True,
                 random_state=0)

xx=[]
yy=[]
for i in X:
    xx.append(i[0])
    yy.append(i[1])

plt.scatter(xx,yy,c='g')

plt.show()#对随机产生的样本进行可视化处理
```

```python
def KMeans(dataset,k):
    m,n=dataset.shape
    cluster=np.zeros(m)   #记录每个样本所属的簇
    
    judge=True   #将每个点的判断设置为1
    center=np.zeros((k,n))
    for i in range(k):
        index=int(np.random.uniform(0,m))
        center[i,:]=dataset[index,:]
    while judge:
        judge=False
        
        #遍历所有样本
        for i in range(m):
            minDist=100000
            minIndex=-1
            
            #遍历所有质心
            for j in range(k):
                dist=np.sqrt(np.sum(np.power(center[j,:]-dataset[i,:] , 2)))#计算欧式距离
                if dist<minDist:
                    minDist=dist
                    minIndex=j
            #更新每个样本所属的簇
            if cluster[i] != minIndex:
                judge=True
                cluster[i]=minIndex
        #更新质心
        for j in range(k):
            points=dataset[cluster==j]
            center[j,:]=np.mean(points,axis=0)
        
    return center,cluster
```
